```python
import pandas as pd
def fetch_quandl_data(symbols,start_date,end_date,price_type):
    df = pd.DataFrame()
    for symbol in symbols:
        
        quandldata = quandl.get("WIKI/{}".format(symbol),start_date=start_date,end_date=end_date)
        quandldata = quandldata[price_type]
        df = pd.concat([df,quandldata],axis=1)
    df.columns = symbols
    df.price_type = price_type
    return df
```

### 5. Black-Scholes Formula

$$Call=S N(d_1) - e^{-r T} K N(d_2)$$
$$Call - Put = df(T) (F-K) = S - e^{-r T} K$$
where
$$d_1 = \frac{log(\frac{S}{K})+rT + \frac{\sigma T^2}{2}}{\sigma \sqrt{T}}$$
$$d_2 = \frac{log(\frac{S}{K})+rT -\frac{\sigma T^2}{2}}{\sigma \sqrt{T}}$$

```python
import scipy as sp
import scipy.stats
import numpy as np

def BlackSholesPrice(S,K,T,r,sigma):
    d1 = (np.log(S/K) + r*T + sigma**2*T/2)/sigma/np.sqrt(T)
    d2 = d1 - sigma * np.sqrt(T)

    call_price = S*sp.stats.norm.cdf(d1) - np.exp(-r*T)*K*sp.stats.norm.cdf(d2)
    put_price = call_price + np.exp(-r*T)*K - S
       
    return call_price,put_price
```

### 6. Implement the Bisection method to nd the root of arbitrary functions.


```python
def bisection(func,interval,tol):
    a,b = interval
    c = (a+b)/2.
    if func(a)*func(b)>0:
        print("warning: fun(a)*fun(b)>0")
#     func(a) - func(b)
    while(b-a>tol):
        if func(c) == 0:
            return c
        elif func(a)*func(c)<0:
            b=c
        else:
            a=c
        c = (a+b)/2.
    return c
```

```python
bisection(np.sin,(-1,1),1e-6)
```

```python
def impliedvol(S,K,T,r,price,option_type):
    if(option_type=="c"):
        price_equation = lambda sigma: BlackSholesPrice(S,K,T,r,sigma)[0] - price
    elif(option_type=="p"):
        price_equation = lambda sigma: BlackSholesPrice(S,K,T,r,sigma)[1] - price
        
    imp_vol = bisection(price_equation,(-1,2),1e-6)
    return imp_vol
```

```python
def calculate_IV_for_quotes(option_df,quote_series,r,option_type):
    S = float(quote_series["regularMarketPrice"])
    t0 = spy_time = int(quote_series["regularMarketTime"])
    t1 = int(option_df["expiration"][0])
#     print(t1)
    T = (t1-t0)/3600/24/365
    
    def calc_iv(row):
        price = (float(row["bid"])+float(row["ask"]))/2
        K = float(row["strike"])
        return impliedvol(S,K,T,r,price,option_type)
    def calc_price(row):
        price = (float(row["bid"])+float(row["ask"]))/2
        return price
        
    option_df["my_iv"] = option_df.apply(calc_iv,axis=1)
    option_df["time_to_maturity"] = np.array([T]*len(option_df))
    option_df["days_count"] = np.array((t1-t0)/3600/24)
    option_df["average_price"] = option_df.apply(calc_price,axis=1)
    option_df["underlying_price"] = [S]*len(option_df)
    option_df["three_months_rate"] = [r]*len(option_df)
    return option_df
```

### Conclusions: 

* Changes in an option's strike price affect whether the option is in-the-money or out-of-the-money. The more an option is in-the-money or out-of-the-money, the greater its implied volatility becomes. The volatility smile may be explained by investor demand for options of the same expiration date but disparate strike prices. In-the-money and out-of-the-money options are usually more desired by investors than at-the-money options. As the price of an option increases all else equal, the implied volatility of the underlying asset increases. Because increased demand bids the prices of such options up, the implied volatility for those options seem to be higher.

* We noticed that there are many contracts that implied volatility tend to zero. This might caused by the illiquidity of that extremely ITM or OTM contracts. Nevertheless, one can also detect that sometimes the ATM implied volatility is higher than trend, especially in put option. This is because people frequently deal these ATM put contracts to hedge their risk, which makes them more appreciable.

### 10. Greeks: Compare numerical method and theoretical formula results for option Greeks.

```python
def numerical_greeks(S,K,T,r,sigma,option_type):
    epsilon=1e-4
    if(option_type=="c"):
        delta = (BlackSholesPrice(S+epsilon,K,T,r,sigma)[0]-BlackSholesPrice(S-epsilon,K,T,r,sigma)[0])/2/epsilon
        vega = (BlackSholesPrice(S,K,T,r,sigma+epsilon)[0]-BlackSholesPrice(S,K,T,r,sigma-epsilon)[0])/2/epsilon
        gamma = (BlackSholesPrice(S+epsilon,K,T,r,sigma)[0]-2*BlackSholesPrice(S,K,T,r,sigma)[0]+\
             BlackSholesPrice(S-epsilon,K,T,r,sigma)[0])/epsilon**2
    elif(option_type=="p"):
        delta = (BlackSholesPrice(S+epsilon,K,T,r,sigma)[1]-BlackSholesPrice(S-epsilon,K,T,r,sigma)[1])/2/epsilon
        vega = (BlackSholesPrice(S,K,T,r,sigma+epsilon)[1]-BlackSholesPrice(S,K,T,r,sigma-epsilon)[1])/2/epsilon
        gamma = (BlackSholesPrice(S+epsilon,K,T,r,sigma)[1]-2*BlackSholesPrice(S,K,T,r,sigma)[1]+\
             BlackSholesPrice(S-epsilon,K,T,r,sigma)[1])/epsilon**2
    return delta,vega,gamma
```

```python
def theoretical_greeks(S,K,T,r,sigma,option_type):
    N = lambda x: sp.stats.norm.cdf(x)
    d1 = (np.log(S/K) + r*T + sigma**2*T/2)/sigma/np.sqrt(T)
    d2 = d1 - sigma * np.sqrt(T)
    
    if(option_type=="c"):
        delta = N(d1)
        gamma = 1/(S*sigma*np.sqrt(T*2*np.pi))*np.exp(-d1**2/2)
        vega = S*np.sqrt(T)*np.exp(-.5*d1**2)/np.sqrt(2*np.pi)
    elif(option_type=="p"):
        delta = (N(d1)-1)
        gamma = 1/(S*sigma*np.sqrt(T*2*np.pi))*np.exp(-d1**2/2)
        vega = S*np.sqrt(T)*np.exp(-.5*d1**2)/np.sqrt(2*np.pi)
    return delta,vega,gamma
```

### 1. Implement the trapezoidal and the Simpson's quadrature rules to numerically approximate the indefinite integral as follows.

$$f(x) = \begin{cases}
\frac{sin(x)}{x} &x \neq 0\\
1 & x=0
\end{cases}
$$

```python
def Simpson(f, a, b, n=500):
    """
    Return the approximation of the integral of f
    from a to b using Simpson's rule with n intervals.
    """
    n=int(n)
    if a > b:
        print('Error: a={} > b={}'.format(a, b))
        return None

    # Check that n is even
    if n % 2 != 0:
#         print('Error: n={} is not an even integer!'.format(n))
        n = n+1  # make n even

    h = (b - a)/float(n)
    sum1 = 0
#     print(n/2+1)
    for i in range(1, int(n/2) + 1):
        sum1 += f(a + (2*i-1)*h)

    sum2 = 0
    for i in range(1, int(n/2)):
        sum2 += f(a + 2*i*h)

    integral = (b-a)/(3*n)*(f(a) + f(b) + 4*sum1 + 2*sum2)
    return integral
```

```python
def Trapezoid(f, a, b, n=500):
    """
    Return the approximation of the integral of f
    from a to b using Simpson's rule with n intervals.
    """
    n=int(n)
    if a > b:
        print('Error: a={} > b={}'.format(a, b))
        return None
    x_array = np.linspace(a,b,n+1)
    col = np.ones(n+1)
    col[0] = 1/2
    col[-1] = 1/2
    dx = (b-a)/n
    return np.dot(f(x_array)*dx,col)
    
```

### 3. In this part, we write a function to decide the integration steps. The algorithm is conducted by an iterative process. In each iteration, we check the absolute error of the recent result and the result one step before. If the absolute error is smaller than some tolerance value, the algorithm stops.

```python
def find_convergence(func, n_list, tol):
    for i,n in enumerate(n_list):
        if(i!=0 and np.abs(func(n)-old)<tol):
            break
        old = func(n)
    if(old != func(n_list[-1])):
        return n,func(n)
    else:
        print("please enlarge your steps")
        
```

### Take $a=10^4$ as an example. We check both the simpson rule and the trapzoid rule for sinc function

```python
trapezoid_n,trapezoid_value = find_convergence(lambda n: Trapezoid(my_sinc,-1e4,1e4,n),np.linspace(5e3,1e5,50),1e-4)
int(trapezoid_n)
```

### Conclusion

It cannot be viewed as a measure of convergency, neither does it be used as a criterion between different integration rules.
