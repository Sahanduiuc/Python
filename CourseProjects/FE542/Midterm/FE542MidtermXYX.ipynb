### 2.1 Suppose that the simple return of a monthly bond index follow the MA(1) model
$$ R_t = a_t + 0.2 a_{t-1},\ \sigma_a=0.025$$
### Assume that $a_{100} = 0.01$. Compute the 1-step- and 2-step-ahead forecasts of the return at the forecast origin t=100. What are the standard deviations of the associated forecast errors? Also compute the lag-1 and lag-2 autocorrelations of the return series.

#### i) Forecasts of return
> $$ \begin{align}
        E[R_{t}| \mathcal{F}_{t-1}] &= E[a_{t}+0.2 a_{t-1}| \mathcal{F}_{t-1}]\\
        &= E[a_{t}| \mathcal{F}_{t-1}] + 0.2 a_{t-1}\\
        &= 0.2 a_{t-1}
     \end{align}$$
> $$ \begin{align}
        E[R_{t}| \mathcal{F}_{t-2}] &= E[a_{t}+0.2 a_{t-1}| \mathcal{F}_{t-2}]\\
        &= E[a_{t}| \mathcal{F}_{t-2}] + 0.2 E[a_{t-1}| \mathcal{F}_{t-2}]\\
        &= 0 + 0.2 \cdot 0\\
        &=0
     \end{align}$$
> Then, by substitution
> $$ E[R_{101}| \mathcal{F}_{100}]= 0.2 a_{100}=0.002$$
> $$ E[R_{102}| \mathcal{F}_{100}]= 0$$  

#### ii) Standard deviation of the associated forecast error
> Definition: $$Std(p) := Std(R_t - E[R_t|\mathcal{F}_{t-p}]$$
> \begin{align}
        Std(1) &= Std(R_t - 0.2 a_{t-1})\\
        &= Std(a_t)\\
        &= \sigma_a\\
        &= 0.025
  \end{align}
>Since $a_t$ and $a_{t-1}$ are independent
>\begin{align}
    Std(2) &= Std(R_t - 0)\\
        &= Std(a_t +0.2 a_{t-1})\\
        &= \sigma_a \sqrt{1+0.2^2}\\
        &= 0.0255
  \end{align}

#### iii) Autocorrelation of return series
>Since this MA(1) process is staionary
$$\begin{align}
    ACF(p)&= \frac{E[(R_t-\mu) (R_{t-p}-\mu)]}{Var[R]}\\
    &= \frac{E[(R_t R_{t-p}]}{Var[R]}
    \end{align}$$
>autocovariance
$$\begin{align}
    E[R_t R_{t-1}]&= E[(a_t + 0.2a_{t-1})(a_{t-1} + 0.2a_{t-2})]\\
    &= E[0.2 a_{t-1}^2] + 0\\
    &= 0.2 \sigma_a^2
   \end{align}$$
$$\begin{align}
    E[R_t R_{t-2}]&= E[(a_t + 0.2a_{t-1})(a_{t-2} + 0.2a_{t-3})]\\
    &= 0
   \end{align}$$
>variance
$$\begin{align}
    Var[R]&=E[(a_t + 0.2 a_{t-1})^2]\\
    &=\sigma_a^2(1+0.2^2)\\
    &=1.04 \sigma_a^2
    \end{align}$$
>Autocorrelations
>$$ACF(1) = \frac{0.2 \sigma_a^2}{1.04 \sigma_a^2} =  0.192$$
>$$ACF(2) = \frac{0}{1.04 \sigma_a^2} =  0$$

### 2.2 $r_t = 0.01 + 0.2 r_{t-2} + a_t$, $a_t \sim \mathcal{N}(0,0.02)$

#### i) What are the mean and variance of the return series $r_t$?

>$$(1-0.2B^2)r_t = 0.01+ a_t$$
> Solve the characteristic polinomial $1-0.2B^2=0$, all of the roots are greater than 1, so the process is stationary
> Then, use the property of stationary process. Add expectation and variance operator to both sides, we get
> $$E[r] = 0.01 + 0.2 E[r] + 0$$
> $$E[r] = 0.01/0.8 = 0.0125 =\mu$$
> Taking the variance of the prior equation, we have
> $$\begin{align}
    Var[r_t] &= E[(0.01+0.2 r_{t-2}+a_t-\mu)^2]\\
    & = E[((0.01+a_t - 0.8 \mu) + 0.2(r_{t-2} - \mu)))^2]\\
    & = E[(0.01 + a_t -0.8 \mu)^2] + 0.2^2 Var[r_{t-2}] + 0.4 E[(r_{t-2} - \mu) (0.01+a_t-0.8 \mu)]\\
    & = (0.01-0.8 \mu)^2 + \sigma_a^2 + 0.2^2 Var[r] +0
   \end{align}
    $$
> $$Var[r] = \frac{(0.01-0.8 \mu)^2 + \sigma_a^2}{1-0.2^2} = 0.0208$$

#### ii)Comput the lag-1 and lag-2 autocorrelations of $r_t$

>lag-1 autocorrelation
$$\begin{align}
    ACF(1) &= \frac{E[(r_t-\mu) (r_{t-1}-\mu)]}{Var[r]}\\

    \end{align}$$
>lag-2 autocorrelation

    ACF(2) &= \frac{E[(r_t-\mu) (r_{t-2}-\mu)]}{Var[r]}\\

```python
import datetime
def preprocess_wrds_data(df):
    df['date'] = df['caldt'].apply(lambda x: datetime.datetime.strptime(str(x),"%Y%m%d"))
    df.index = df.date
    df.drop(['date'],axis=1,inplace=True)
    df['simple_return'] = (df.totval-df.totval.shift(1))/df.totval.shift(1)
    df['log_return'] = np.log(df.totval/df.totval.shift(1))
    df.dropna(axis=0,how='any',inplace=True)
```

### (b)  Study unit root non-stationarity.  Compare for all data periods.

> Unit-root checking

```python
from statsmodels.tsa.stattools import adfuller

def unit_root_cheking(time_series,regression):  
    # perform Dickey-Fuller test
    dftest = adfuller(time_series,regression='c',autolag="BIC")
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value',
                                             '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
for i in range(7):
    print("data",i+1)
    unit_root_cheking(data[i],'c')
```

> Conclusion: All results reject the unit-roots. Because we are considering the first difference of the price itself it is much more stationary.

> Model screening. Zero indicate there is an error in training data. Unfortunately it seems that in Python's statsmodels library, we cannot use ARMA(0,0) properly. It's not a big deal though.

```python
bic_matrix = np.zeros((7,3,3))
for i in range(7):
    for j in range(3):
        for k in range(3):
            try:
                model=sm.tsa.ARMA(data[i],(j,k)).fit(trend='nc')
                bic_matrix[i,j,k] = model.bic
            except:
                pass
            
print(bic_matrix)
```

```python
def find_min_idx(a):
    p = np.argmin(a)
    return (p//a.shape[1],p%a.shape[1])

def model_screening(bic_matrix):
    ar_model = []
    ma_model = []
    arma_model = []
    ndata = bic_matrix.shape[0]
    for i in range(ndata):
        best_ar_order = np.argmin(bic_matrix[i][:][0])
        ar_model.append(sm.tsa.ARMA(data[i],(best_ar_order,0)).fit(trend='nc'))
        best_ma_order = np.argmin(bic_matrix[i][0][:])
        ma_model.append(sm.tsa.ARMA(data[i],(0,best_ma_order)).fit(trend='nc'))
        best_arma_order = find_min_idx(bic_matrix[i])
        arma_model.append(sm.tsa.ARMA(data[i],best_arma_order).fit(trend='nc'))
    return (ar_model,ma_model,arma_model)
```

```python
(ar_model,ma_model,arma_model) = model_screening(bic_matrix)
```
