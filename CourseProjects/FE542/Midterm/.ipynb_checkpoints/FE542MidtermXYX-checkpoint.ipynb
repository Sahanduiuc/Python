### (a)  Work  with  continuously  compounded  returns.   Pay  attention  tothe proper date order in the data you downloaded.

>Preprocessing

```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
import datetime
def preprocess_wrds_data(df):
    df['date'] = df['caldt'].apply(lambda x: datetime.datetime.strptime(str(x),"%Y%m%d"))
    df.index = df.date
    df.drop(['date'],axis=1,inplace=True)
    df['simple_return'] = (df.totval-df.totval.shift(1))/df.totval.shift(1)
    df['log_return'] = np.log(df.totval/df.totval.shift(1))
    df.dropna(axis=0,how='any',inplace=True)
```

```python
df = pd.read_csv('CRSP_on_SP500_20y.csv')
print(df.head())
preprocess_wrds_data(df)
print(df.tail())
```

### (b)  Study unit root non-stationarity.  Compare for all data periods.

> Unit-root checking

```python
from statsmodels.tsa.stattools import adfuller

def unit_root_cheking(time_series,regression):  
    # perform Dickey-Fuller test
    dftest = adfuller(time_series,regression='c',autolag="BIC")
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value',
                                             '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
for i in range(7):
    print("data",i+1)
    unit_root_cheking(data[i],'c')
```

> Conclusion: All results reject the unit-roots. Because we are considering the first difference of the price itself it is much more stationary.

### (c)  Build the best AR, the best MA and the best ARMA model afteryou remove non-stationarity (if needed) for each time period.

> ACF and PACF

```python
plt.figure(figsize=(12,20))
for i in range(7):
    ax1 = plt.subplot(7,2,2*i+1)
    sm.graphics.tsa.plot_acf(data[i], alpha=0.05, lags=40, ax=ax1)
    ax2 = plt.subplot(7,2,2*i+2)
    sm.graphics.tsa.plot_pacf(data[i], alpha=0.05, lags=40, ax=ax2)
    
plt.show()
```

> Model screening. Zero indicate there is an error in training data. Unfortunately it seems that in Python's statsmodels library, we cannot use ARMA(0,0) properly. It's not a big deal though.

```python
bic_matrix = np.zeros((7,3,3))
for i in range(7):
    for j in range(3):
        for k in range(3):
            try:
                model=sm.tsa.ARMA(data[i],(j,k)).fit(trend='nc')
                bic_matrix[i,j,k] = model.bic
            except:
                pass
            
print(bic_matrix)
```

```python
def find_min_idx(a):
    p = np.argmin(a)
    return (p//a.shape[1],p%a.shape[1])

def model_screening(bic_matrix):
    ar_model = []
    ma_model = []
    arma_model = []
    ndata = bic_matrix.shape[0]
    for i in range(ndata):
        best_ar_order = np.argmin(bic_matrix[i][:][0])
        ar_model.append(sm.tsa.ARMA(data[i],(best_ar_order,0)).fit(trend='nc'))
        best_ma_order = np.argmin(bic_matrix[i][0][:])
        ma_model.append(sm.tsa.ARMA(data[i],(0,best_ma_order)).fit(trend='nc'))
        best_arma_order = find_min_idx(bic_matrix[i])
        arma_model.append(sm.tsa.ARMA(data[i],best_arma_order).fit(trend='nc'))
    return (ar_model,ma_model,arma_model)
```

```python
(ar_model,ma_model,arma_model) = model_screening(bic_matrix)
```

>Best Models Output

```python
for i in range(7):
    print("*******************************************************")
    print("*******************************************************")
    print("The best ARMA model of data", i+1)
    print(arma_model[i].params)
    print("The best AR model of data", i+1)
    print(ar_model[i].params)
    print("The best MA model of data", i+1)
    print(ma_model[i].params)
```

### (d)  Build a seasonal model for your data using whatever lag you feelapproximates the data best.

#### i) Experiment  (One can skip this section, it's <span style="color:red">not </span> relevant to the answer)

> Experiment: Cosine period data
> This is an independent problem, one can skip this section
> Assume we know $$y_k = cos(\frac{k \pi}{4} ) + e_k$$
> where $$e_k \sim N(0,1)$$
> Apparently there is a period 8 under $y_k$, should we construct a seasonal model?

```python
x = np.cos([i*0.25*np.pi for i in range(1000)])
e = sp.stats.norm(0,1).rvs(1000)
y = x+e
y = pd.DataFrame(y)
plt.plot(y)
plt.show()
```

> Plot ACF, one can see there is an obvious bump at lag=8

```python
plt.figure(figsize=(12,12))
plt.subplot(5,1,1)
plt.plot(y)
ax2 = plt.subplot(5,1,2)
sm.graphics.tsa.plot_acf((y-y.shift(8)).dropna(axis=0,inplace=False),alpha=0.05,lags=20,ax=ax2)
plt.show()
```

> ARMA(1,8) model with the specific non-zero 8th MA lag

```python
x = np.cos([i*0.25*np.pi for i in range(1000)])
e = sp.stats.norm(0,1).rvs(1000)
y = x+e
ma_lag = np.zeros(8)
ma_lag[7]=1
mod = sm.tsa.SARIMAX(y,order=(1,0,ma_lag))
cos_arma_model = mod.fit()
print(cos_arma_model.summary())
```

> ARMAX(1,0,0) model with hidden process $x_k = cos(\frac{k \pi}{4} )$

```python
x = np.cos([i*0.25*np.pi for i in range(1000)])
e = sp.stats.norm(0,1).rvs(1000)
y = x+e
mod = sm.tsa.SARIMAX(y,x,order=(1,0,0))
cos_armax_model = mod.fit()
print(cos_armax_model.summary())
```

```python
x = np.cos([i*0.25*np.pi for i in range(1000)])
e = sp.stats.norm(0,1).rvs(1000)
y = x+e
cos_sarima_model = sm.tsa.SARIMAX(y,order=(1,0,0),seasonal_order=(1,1,1,8)).fit()
print(cos_sarima_model.summary())
```

```python
print("arma",cos_arma_model.bic)
print("armax",cos_armax_model.bic)
print("sarima",cos_sarima_model.bic)
```

>Conclusion: If you know the hidden process x, armax should be most effective. However, in practice, we are not supposed to know them, seasonal arima model should be an good alternative

#### iii) Daily data analysis (This is the answer)

> However, if we use the daily data, seasonality is not always obvious. Here I use SARIMAX model with hidden process and seasonal orders. The hidden process is set to be the annual moving average process, and the seansonal order=(0,1,0,252)

```python
seasonal_period=252
seasonal_mod = sm.tsa.SARIMAX(data[4][seasonal_period-1:],data[4].rolling(seasonal_period).mean()[seasonal_period-1:],
                     order=(1,0,1),seasonal_order=(0,1,0,seasonal_period))
seasonal_res = seasonal_mod.fit(disp=True)
print(seasonal_res.summary())
```

```python
print("ar model BIC:",ar_model[3].bic)
print("ma model BIC:",ma_model[3].bic)
print("arma model BIC:",arma_model[3].bic)
print("seasonal model BIC:",seasonal_res.bic)
```

> Conclusion: 
> Does the S&P500 data has seasonality? Yes, but only obvious in monthly(or rougher sampling frequency) data.
> The BIC of the seasonal model is much greater than AR, MA, ARMA model.

>Following, we should **not** consider seasonality to the daily data.
